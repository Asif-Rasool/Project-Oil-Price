# Checking stationarity of each variable: nm_oil_price

## Creating a subset to run statotianrity tests

```python
nm_oil_price = df[['nm_oil_price']].copy()
nm_oil_price
```

## Visualizing nm_oil_price

```python
nm_oil_price.plot()
```

## ADF test for stationarity

```python
test_result=adfuller(nm_oil_price['nm_oil_price'])

#Ho: It is non stationary
#H1: It is stationary

def adfuller_test(nm_oil_price):
    result=adfuller(nm_oil_price)
    labels = ['ADF Test Statistic','p-value','#Lags Used','Number of Observations Used']
    for value,label in zip(result,labels):
        print(label+' : '+str(value) )
    if result[1] <= 0.05:
        print("strong evidence against the null hypothesis(Ho), reject the null hypothesis. Data has no unit root and is stationary")
    else:
        print("weak evidence against null hypothesis, time series has a unit root, indicating it is non-stationary ")
        
adfuller_test(nm_oil_price['nm_oil_price']) 
```

## Differencing to correct stationarity

```python
nm_oil_price['nm_oil_price_sead'] = nm_oil_price['nm_oil_price'] - nm_oil_price['nm_oil_price'].shift(1)
```

## Plotting after taking first differencing 

```python
plt.plot(nm_oil_price['nm_oil_price'], label = 'non-stationary')
plt.plot(nm_oil_price['nm_oil_price_sead'], label = 'stationary')
plt.title('nm_oil_price_sead')
plt.legend()
plt.savefig(r'C:\Users\asifr\OneDrive - State of New Mexico\Documents\Project Oil Price\temp\image0.png')
```

## Rerunning ADF test to check if non-stationarity was corrected 

```python
nm_oil_price = nm_oil_price.dropna()

from statsmodels.tsa.stattools import adfuller

test_result=adfuller(nm_oil_price['nm_oil_price_sead'])

#Ho: It is non stationary
#H1: It is stationary

def adfuller_test(nm_oil_price_sead):
    result=adfuller(nm_oil_price_sead)
    labels = ['ADF Test Statistic','p-value','#Lags Used','Number of Observations Used']
    for value,label in zip(result,labels):
        print(label+' : '+str(value) )
    if result[1] <= 0.05:
        print("strong evidence against the null hypothesis(Ho), reject the null hypothesis. Data has no unit root and is stationary")
    else:
        print("weak evidence against null hypothesis, time series has a unit root, indicating it is non-stationary")
        
adfuller_test(nm_oil_price['nm_oil_price_sead']) 
```

## Exporting to an excel file

```python
nm_oil_price.index = nm_oil_price.index.date

# Create a Pandas Excel writer using XlsxWriter as the engine.
writer = pd.ExcelWriter(r'C:\Users\asifr\OneDrive - State of New Mexico\Documents\Project Oil Price\Outputs\pred_sead_n_plots.xlsx', engine='xlsxwriter')

# Convert the dataframe to an XlsxWriter Excel object.
nm_oil_price.to_excel(writer, sheet_name='nm_oil_price')

# Get the xlsxwriter workbook and worksheet objects.
workbook  = writer.book
worksheet = writer.sheets['nm_oil_price']

# Insert an image.
worksheet.insert_image('F2', r'C:\Users\asifr\OneDrive - State of New Mexico\Documents\Project Oil Price\temp\image0.png')
```
