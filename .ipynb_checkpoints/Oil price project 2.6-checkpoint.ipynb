# Loading data

```python
df = pd.DataFrame(pd.read_excel(r'C:\Users\asifr\OneDrive - State of New Mexico\Documents\Project Oil Price\Outputs\df.xlsx'))
df_pred = pd.DataFrame(pd.read_excel(r'C:\Users\asifr\OneDrive - State of New Mexico\Documents\Project Oil Price\Outputs\df_pred.xlsx'))
pd.options.display.max_columns = 500
df = df.set_index('Date')
df.index = pd.to_datetime(df.index)
df_pred = df_pred.set_index('Date')
df_pred.index = pd.to_datetime(df_pred.index)
```

## Dropping non-regressors

```python
df.drop(['wti_spot_price', 'brent_spot_price',
       'dubai_price', 'index_china_gdp', 'index_us_gdp', 'index_cboe_volt',
       'us_unemploy', 'index_ip_mining', 'index_cpi_china', 'opec', 'non_opec',
       'index_cpi_us', 'index_energy', 'index_metal', 'index_precious',
       'del_supply_nonopec', 'del_cap_per', 'del_world_gdp_per',
       'del_supply_saudi', 'opec_spare_capacity', 'oecd_inventory',
       'del_world_consump'], axis = 1, inplace = True) # we set axis = 1 to remove columns, axis = 0 to remove rows. And inplace means we are modifying our dataframe, 
                                    # we ae not making a copy
```

```python
df_nm = df[['nm_oil_price']].copy(deep= True)
def create_features(df_nm):
    """
    Create time series features based on time series index.
    """
    df_nm = df_nm.copy()
    df_nm['quarter'] = df_nm.index.quarter
    df_nm['year'] = df_nm.index.year
    return df_nm

df_nm = create_features(df_nm)
```

## Creating final dataset to run and predict oil price

```python
df_model = pd.concat([df, df_pred])
```

```python
df_model
```

# Creating train-test split

## Plotting the train-test split

```python
train = df_model.loc[df_model.index < '01-01-2020']
test = df_model.loc[df_model.index >= '01-01-2020']

fig, ax = plt.subplots(figsize=(15, 5))
train["nm_oil_price"].plot(ax=ax, label='Training Set', title='Data Train/Test Split')
test["nm_oil_price"].plot(ax=ax, label='Test Set')
ax.axvline('01-01-2020', color='black', ls='--')
ax.legend(['Training Set', 'Test Set'])
plt.show()
```

```python
df_model.tail()
```

## Creating model features

```python
train = create_features(train)
test = create_features(test)

FEATURES = ['wti_spot_price_sead', 'brent_spot_price_sead', 'dubai_price_sead',\
       'index_china_gdp_sead', 'index_us_gdp_sead', 'index_cboe_volt_sead',\
       'us_unemploy_sead', 'index_ip_mining_sead', 'index_cpi_china_sead',\
       'opec_sead', 'non_opec_sead', 'index_cpi_us_sea', 'index_energy_sead',\
       'index_metal_sead', 'index_precious_sead', 'del_supply_nonopec_sead',\
       'del_cap_per_sead', 'del_world_gdp_per_sead', 'del_supply_saudi_sead',\
       'opec_spare_capacity_sead', 'oecd_inventory_sead',\
       'del_world_consump_sead', 'sv_911', 'sv_covid19', 'sv_gfc', 'sv_lpc',\
       'sv_opec']

TARGET = 'nm_oil_price'

X_train = train[FEATURES]
y_train = train[TARGET]

X_test = test[FEATURES]
y_test = test[TARGET]
```

# Running XGBoost regression

## Running regreesion on train data (entire dataset till year 2020)

```python
reg = xgb.XGBRegressor(base_score=0.5, booster='gbtree',    
                       n_estimators=6000,
                       early_stopping_rounds=50,
                       objective='reg:squarederror',
                       max_depth=3,
                       learning_rate=0.01)
reg.fit(X_train, y_train,
        eval_set=[(X_train, y_train), (X_test, y_test)],
        verbose=100)
```

## Plotting variable importance

```python
fi = pd.DataFrame(data=reg.feature_importances_, 
             index=reg.feature_names_in_,
             columns=['importance'])
fi.sort_values('importance').plot(kind='barh', figsize=(16, 10), title='Feature Importance')
plt.show()
```

## Fitting the model with the predicted regressors

```python
test['prediction'] = reg.predict(X_test)
df_model = df_model.merge(test[['prediction']], how='left', left_index=True, right_index=True)
ax = df_model[['nm_oil_price']].plot(figsize=(15, 5))
df_model['prediction'].plot(ax=ax, style='.')
plt.legend(['True Data', 'Predictions'])
ax.set_title('Raw Data and Prediction')
plt.show()
#df.to_csv('file_name_1.csv')
```
